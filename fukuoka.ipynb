```python
url = 'https://ckan.open-governmentdata.org/dataset/8a9688c2-7b9f-4347-ad6e-de3b339ef740/resource/9d32b7ee-5bfe-4b3c-a582-ea56f3b0afd9/download/400009_pref_fukuoka_covid19_patients.csv'
#https://ckan.open-governmentdata.org/dataset/8a9688c2-7b9f-4347-ad6e-de3b339ef740/resource/0430a12e-568c-4a6a-bed8-51621f47c6e5/download/400009_pref_fukuoka_covid19_patients.csv'
```

```python
df = pd.read_csv(url) #福岡市オープンデータサイトにアクセスし，データを読み取ってデータフレイムに変換
df['公表_年月日'] = df['公表_年月日'].str.replace('/', '-') #2020/02/20 ==>2020-02-20
df
```

```python
today = datetime.date.today() #今日の日付を取得
oneday=datetime.timedelta(days=1) #"1日"を取得
seven = datetime.timedelta(days=7)
yesterday=today-oneday #今日の日付から"1日"引く==>昨日の日付を取得
lastweek = yesterday-seven
yesterday = yesterday.strftime("%Y-%m-%d") #昨日の日付を文字列として取得
lastweek = lastweek.strftime("%Y-%m-%d") #昨日の先週の日付けを文字列として取得
dfa = df[df.公表_年月日 == yesterday] #昨日の日付だけの部分をDataFrameから抜き出し，新たに作成
dfb = df[df.公表_年月日 == lastweek] #先週の日付だけの部分をDataFrameから抜き出し，新たに作成
dfa
```

```python
df2 = dfa["居住地"].value_counts() #同じ市，郡の出現回数をカウント(pandas.Series)
df3 = dfb["居住地"].value_counts()
df2 = pd.DataFrame(df2) #DataFrame型に変換
df3 = pd.DataFrame(df3)
df2
```

```python
df2.columns=['number'] #列ラベルの追加
df2 = df2.rename_axis('residence').reset_index() #index列名を変更
df3.columns=['number'] #列ラベルの追加
df3 = df3.rename_axis('residence').reset_index() #index列名を変更
df2
```

```python
sum = df2["number"].sum() #昨日の新規感染者数数を取得
sum2 = df3["number"].sum() #先週の新規感染者数数を取得
print(sum)
```

```python
if sum-sum2>0: #先週より感染者数が多かったら，先週比の数字に+を付ける
  ratio = f"+{sum-sum2}"
else:
  ratio = sum-sum2
```

```python
df2['residence'] = df2['residence'].str.replace('福岡市', '') #福岡市東区==>東区
df2['residence'] = df2['residence'].str.replace('北九州市', '') #北九州市小倉南区==>小倉南区
```

```python
combDf = fukuokaShp.merge(df2,left_on='N03_004',right_on='residence',how='left') #地図データと感染者数のデータをもつDataFrameを結合
combDf
```

```python
ax = combDf.plot(figsize=(14, 14)) #地図の外形作成
ax.text(x=131.2-0.025, y=34, s='N', fontsize=30) # 地図中に"N"を描画
ax.arrow(131.2, 33, 0, 1, length_includes_head=True,
          head_width=0.06, head_length=0.1, overhang=.1, facecolor='k') # 地図中にNorth Arrowを描画
combDf.apply(lambda x: ax.annotate(s=x.N03_004, xy=x.geometry.centroid.coords[0],
                                   ha='center', color = 'black', size = 6),axis=1) ##地図の中に注略(市，郡，区)を入れる
combDf.plot(column='number', cmap = 'rainbow', edgecolors='black',
            ax = ax, legend=True,legend_kwds={'label': "Number of covid19 infections",'orientation':"vertical"}) #凡例の作成
plt.title('Number of covid19 infections', fontsize=16) #図のタイトルを作成
plt.savefig("img.png") #作成した図を保存
```
